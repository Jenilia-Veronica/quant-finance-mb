Simple Interest Calculator(principal is 1000, annual interest rate is 0.05, time in years is 3)

```python
p = 1000
r = 0.05
t = 3
fv_simple = p*(1+r*t)
interest = fv_simple-p
print("future value: ", fv_simple)
print("interest: ", interest)
```

Continuous Compounding


```python
import numpy as np

p = 1000
r=0.05
t=3

fv = p* np.exp(r*t)
fv_daily=p*(1+r/365)**(365*t)
difference = fv-fv_daily

print("Continuous Compounding: ", f"{fv:.4f}")
print("Daily Compounding: ", f"{fv_daily:.4f}")
```

Calculating PV and Discount Factor for discrete and continuous compounding

```python
import numpy as np

def df(r, t, continuous= False):
    if continuous:
        return np.exp(-r*t)
    return (1+r)**(-t)

fv= 1000
r=0.06
t=5

pv_discrete= fv*df(r, t, continuous=False)
pv_continuous= fv*df(r, t, continuous=True)

print("future value: ", fv, " at", r, " in", t, " years")
print("discount rate: ", r*100, "%")
print("present value(discrete): ₹", f"{pv_discrete:.2f}")
print("present value(continuous): ₹", f"{pv_continuous:.2f}")

print("discount factor(discrete): ", f"{pv_discrete/fv:.6f}")
print("discount factor(continuous): ", f"{pv_continuous/fv:.6f}")
```
